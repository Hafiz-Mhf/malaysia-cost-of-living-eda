```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for a professional look
sns.barplot(
    data=df_latest, 
    x='real_median_income', 
    y='state', 
    hue='state',      # Assign the y-variable to hue
    palette='viridis', 
    legend=False      # Hide the legend since the y-axis already labels the states
)

df = pd.read_csv("../data/cleaned/malaysia_living_costs_cleaned.csv")
```

# Which states face the highest stress?

```python
# Filter for the most recent year in the data
latest_year = df['year'].max()
df_latest = df[df['year'] == latest_year].sort_values('real_median_income', ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(data=df_latest, x='real_median_income', y='state', palette='viridis')
plt.title(f'Real Median Income by State in {latest_year} (Inflation Adjusted)')
plt.xlabel('Adjusted Income (RM)')
plt.ylabel('State')

# Save the plot to your visuals folder
plt.savefig("../visuals/income_by_state_2022.png", dpi=300, bbox_inches='tight')

# Always call plt.show() AFTER savefig
plt.show()
```

# "Income vs. Inflation"

```python
# List of states to compare
target_states = ['Selangor', 'W.P. Kuala Lumpur', 'Kelantan', 'Johor']
df_trends = df[df['state'].isin(target_states)]

plt.figure(figsize=(12, 6))
sns.lineplot(data=df_trends, x='year', y='real_median_income', hue='state', marker='o')
plt.title('Real Median Income Trends: Urban vs. Rural States')
plt.ylabel('Inflation-Adjusted Median Income (RM)')

# Save the plot to your visuals folder
plt.savefig("../visuals/real_income_trends.png.png", dpi=300, bbox_inches='tight')

# Always call plt.show() AFTER savefig
plt.show()
```
